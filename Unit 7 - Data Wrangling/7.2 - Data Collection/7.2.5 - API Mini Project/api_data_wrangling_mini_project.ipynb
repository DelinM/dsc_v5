## 1 Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).

```python
# get url
url_time_2017 = 'start_date=2017-01-02&end_date=2017-12-29'
url_2017 = url + url_stock_set + url_api_key + '&'+ url_time_2017

# get request and json data
afx_stock_2017 = requests.get(url_2017)

afx_stock_2017
```

## 2 Convert the returned JSON object into a Python dictionary.

```python
data = afx_stock_2017.json()
data
```

## 3 Calculate what the highest and lowest opening prices were for the stock in this period.

```python
stock_price = data['dataset']['data']
stock_price
```

```python
stock_price_header = data['dataset']['column_names']
stock_price_header
```

From my observation, stock_price is a list of lists. In each list item, there is a list including the information (in order) included in stock_price_header. Based on the questions of this assignment, we will need data in columns **Open**, **High**, **Low**, and **Traded Volume**.

I will create a dictionary of list that contains the data in the above-mentioned columns. The keys will be dates, and the value for the key would be a list: [value of high, value of low, value of Trade Volum]

```python
# identify the index for "Date", "High", "Low", and "Traded Volumes"
date_index = stock_price_header.index('Date')
open_index = stock_price_header.index('Open')
high_index = stock_price_header.index('High')
low_index = stock_price_header.index('Low')
close_index = stock_price_header.index('Close')
tradedv_index = stock_price_header.index('Traded Volume')
```

```python
# create a new dict to store data
stock_dic = {}

for info in data['dataset']['data']:
    stock_dic[info[date_index]] = [info[open_index], info[high_index], info[low_index], info[close_index], info[tradedv_index]]

stock_dic
```

```python
# use a for loop to find the highest and lowest opening price

highest_openning = 0
lowest_opening = stock_dic['2017-12-29'][0]
highest_date = ''
lowest_date = ''

for date, values in stock_dic.items():
    
    if values[0]:
        if values[0] > highest_openning:
            highest_openning = values[0]
            highest_date = date

        if values[0] < lowest_opening:
            lowest_opening = values[0]
            lowest_date = date

    
print('On the date {}, AFX had its highest openning {} in 2017'.format(highest_date, highest_openning))
print('On the date {}, AFX had its lowest openning {} in 2017'.format(lowest_date, lowest_opening))
```

As mentioned above, the highest opening for AFX in 2017 is \\$53.11, and the lowest is \\$ 34.0.

## 4 What is the largest change in any one day (based on High and Low price)?

```python
largest_change = max(x[1]-x[2] for key,x in stock_dic.items())
largest_change
```

The largest change (based on high and low price) is \\$2.81.

## 5 What was the largest change between any two days (based on Closing Price)?

```python
closing_list = [x[3] for key,x in stock_dic.items()]

largest_difference = max([closing_list[i+1]-closing_list[i] for i in range(len(closing_list)-1)])
largest_difference

largest_difference
```

The largest change (based on closing price) is \\$2.56.

## 6 What was the average daily trading volume during this year?

```python
sum_volume = sum([x[-1] for key,x in stock_dic.items()])
no_date = len(stock_dic)
average = sum_volume/no_date
average
```

The average daily trading volume in 2017 is 89,124.34.

## 7 (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
volume_list = [x[-1] for key,x in stock_dic.items()]
volume_list = sorted(volume_list)

volume_len = len(volume_list)

if volume_len%2 ==0:
    median_len = volume_len/2
else:
    median_len = round(volume_len/2)

    
# for index, i need to mins 1 from the location of the median
median_volume = volume_list[median_len -1]
median_volume
```

The medium trading volume in 2017  was 76,286.0.
